```python
from baselines.bench.monitor import load_results
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
a2c_plain_results_df = load_results('runs/A2C_FramePredictRNNSim_debugging-plain-with-wrapper_20180814173807/a2c')
a2c_curious_point1_results_df = load_results('runs/A2C_FramePredictRNNSim_now_working_0.1_curiosity_20180815023409/a2c')
a2c_curious_point01_results_df = load_results('runs/A2C_FramePredictRNNSim_now_working_0.01_curiosity_20180815023454/a2c')
a2c_point1_curious_point05_entropy_results_df = load_results('runs/A2C_FramePredictRNNSim_now_working_0.1_curiosity_0.05_entropy_20180815164042/a2c')
a2c_plain_point_05_entropy_results_df = load_results('runs/A2C_FramePredictRNNSim_now_working_plain_0.05_entropy_20180815164027/a2c')
```

```python
a2c_plain_results_df.describe()
```

```python
# Average results among all the parallel processes.
a2c_plain_results_df = a2c_plain_results_df.groupby('index').mean().reset_index()
a2c_curious_point1_results_df = a2c_curious_point1_results_df.groupby('index').mean().reset_index()
a2c_curious_point01_results_df = a2c_curious_point01_results_df.groupby('index').mean().reset_index()
a2c_point1_curious_point05_entropy_results_df = a2c_point1_curious_point05_entropy_results_df.groupby('index').mean().reset_index()
a2c_plain_point_05_entropy_results_df = a2c_plain_point_05_entropy_results_df.groupby('index').mean().reset_index()
```

```python
a2c_plain_results_df.plot.line(x='index',y='r',title='normal entropy 0.1')
a2c_curious_point1_results_df.plot.line(x='index',y='r',title='curious 0.1 entropy 0.1')
a2c_curious_point01_results_df.plot.line(x='index',y='r',title='curious 0.01 entropy 0.1')
a2c_point1_curious_point05_entropy_results_df.plot.line(x='index',y='r',title='curious 0.1 entropy 0.05')
a2c_plain_point_05_entropy_results_df.plot.line(x='index',y='r',title='normal entropy 0.05')

plt.show()
```
